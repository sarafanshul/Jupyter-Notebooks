```python
import pandas as ps
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
time = np.linspace(0, 4*np.pi, 200)
```

```python
plt.figure(figsize =(15 , 4) , dpi = 80 , facecolor = 'w' )

plt.plot(time, np.sin(time) , label='sin(x)')
plt.plot(time, np.sin(time - 2*(np.pi)/3) , color='r' ,label='sin(x+pi/3)')
plt.plot(time, np.sin(time + 2*(np.pi)/3) , color = 'g' , label='sin(x+2pi3)')

plt.title('3-Phase-AC')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.axhline(y=0, color='k')
plt.axvline(x = np.pi/2 , color = '#222f3e' , linestyle = '--' )
plt.legend(loc = 'lower right')
```

```python
x = np.linspace(0, 4*np.pi, 500)
```

```python
plt.figure(figsize =(15 , 5))
plt.plot(x, np.sin(x**2), '-',label='sin(x^2)')
plt.plot(x, np.sin(x) , '-r' , label='sin(x)')
# plt.legend(loc='upper left')
plt.legend()
plt.grid(None)
plt.axvline(x=np.pi/2, color = '#222f3e' , linestyle = '--' )
plt.title('A simple chirp');
```
