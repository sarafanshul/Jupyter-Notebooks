```python
plt.style.use('seaborn-deep')
plt.figure( figsize = (18 , 7) , dpi = 80 , facecolor='w' ,edgecolor = 'k')

plt.stackplot( minutes , pl_1 , pl_2 , pl_3  ,labels = labels)
plt.legend(loc = 'upper left')
plt.grid(True)
plt.title('StackPlot')
# plt.tight_layout()
```

## Histogram

```python
import statistics 
data = pd.read_csv(r'DATA/datahist.csv')
ids = data['Responder_id']
ages = data['Age']
bins = list(range(10 , 81 , 10))
median_age = statistics.median(ages)
```

```python
plt.style.use('seaborn-deep')
plt.figure( figsize = (10 , 7) , dpi = 80 , facecolor='w' ,edgecolor = 'k')

plt.hist( ages , bins = bins ,color = '#00b894', alpha = 0.5 )

plt.axvline(median_age , color = '#d63031' , alpha = 1 , label = 'Age Median')
plt.title('Histogram')
plt.legend()
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Ages',fontsize=15)
plt.ylabel('Frequency',fontsize=15)
```

```python
from matplotlib import colors
from matplotlib.ticker import PercentFormatter

# Fixing random state for reproducibility
np.random.seed(19680801)
N_points = 100000
n_bins = 20

# Generate a normal distribution, center at x=0 and y=5
x = np.random.randn(N_points)
y = .4 * x + np.random.randn(100000) + 5

plt.figure( figsize = (10 , 7) , dpi = 80 , facecolor='w' ,edgecolor = 'k')

# N is the count in each bin, bins is the lower-limit of the bin
N, bins, patches = plt.hist(x, bins=n_bins)

# We'll color code by height, but you could use any scalar
fracs = N / N.max()

# we need to normalize the data to 0..1 for the full range of the colormap
norm = colors.Normalize(fracs.min(), fracs.max())

# Now, we'll loop through our objects and set the color of each accordingly
for thisfrac, thispatch in zip(fracs, patches):
    color = plt.cm.viridis(norm(thisfrac))
    thispatch.set_facecolor(color)
```

```python
plt.style.use('seaborn-deep')
plt.figure( figsize = (10 , 7) , dpi = 80 , facecolor='w' ,edgecolor = 'k')

plt.hist2d(x , y , bins=100 )

plt.title('2D Histograph')
```
