```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
data = pd.read_csv(r'DATA/weather.csv')
```

```python
data_droped = data.drop(['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC','JF','MAM','JJAS','OND'], axis=1)
group = data_droped.groupby('SUBDIVISION')
```

```python
for g , g_df in group :
    print(g,type(g))
    print(g_df.head(),f'Type = {type(g_df)} , Length = {len(g_df)}')
```

```python
# plt.bar( group.YEAR , group.ANNUAL.max())
%matplotlib inline
group.plot()
```

```python
#Sorting Out Usefull Data
data_mp = data[data.SUBDIVISION == 'West Madhya Pradesh']
data_mh = data[data.SUBDIVISION == 'Madhya Maharashtra']
data_gj = data[data.SUBDIVISION == 'Gujarat Region']
```

```python
#-------INITIALING THE WINDOW---------
plt.style.use('seaborn-deep')
plt.figure(figsize =(15 , 6) ,  facecolor='w', edgecolor='k')
plt.grid()

plt.plot(data_mp.YEAR , data_mp.ANNUAL, color = '#222f3e' , linestyle = '-' , marker = 'o' ,linewidth=2 ,label='Madhya Pradesh')
plt.plot(data_mh.YEAR , data_mh.ANNUAL, color = '#ee5253' , linestyle = '--' , marker = '.' ,linewidth=1.3 ,label='Maharastra')
plt.plot(data_gj.YEAR , data_gj.ANNUAL, color = '#0abde3' , linestyle = '--' , marker = '.' ,linewidth=1.3 ,label='Gujarat')
plt.legend()
#---------COMMON ATTRIBUTES-----------
plt.xlabel('Year')
plt.ylabel('Median Rianfall')
plt.title('Comparision of Median Rainfall in Madhya Pradesh , Maharastra , Gujarat by Year')
#plt.plot()
```

```python
#Sorting Out Usefull Data
mp = data_mp[data_mp.YEAR != 2000]
mh = data_mh[data_mh.YEAR != 2000]
gj = data_gj[data_gj.YEAR != 2000]
time = mh.YEAR
labels = ['Madhya Pradesh' , 'Maharastra' , 'Gujarat']
color = ['#2ecc71','#e74c3c','#f1c40f']
```

```python
#-------INITIALING THE WINDOW---------
plt.figure(figsize=(18, 10), dpi= 80)
plt.style.use('seaborn-deep')

plt.stackplot( time , mp.ANNUAL , mh.ANNUAL , gj.ANNUAL , colors = color , labels = labels)

#-------------------Styling------------------
plt.legend(loc='upper left')
plt.title('Comparision of Median Rainfall in Madhya Pradesh , Maharastra , Gujarat by Year')
```

```python
#Sorting Out Usefull Data
data_mp_bar = data_mp[data_mp.YEAR > 2000]
data_mh_bar = data_mh[data_mh.YEAR > 2000]
data_gj_bar = data_gj[data_gj.YEAR > 2000]
x_indexes = np.arange(len(data_mp_bar.YEAR))
width = 0.25
```

```python
#-------INITIALING THE WINDOW---------
plt.figure(figsize=(18, 10), dpi= 80)
plt.style.use('seaborn-white')

plt.bar(x_indexes + width , data_mp_bar.ANNUAL , width=width , label='Madhya Pradesh' )
plt.bar(x_indexes  , data_mh_bar.ANNUAL , width=width ,  label = 'Maharatra' ) 
plt.bar(x_indexes - width , data_gj_bar.ANNUAL , width=width , label = 'Gujarat' )

#changing indexes with Year
plt.xticks( ticks = x_indexes , labels = data_mp_bar.YEAR )

#-------------------Styling------------------
plt.legend(loc='upper left')
plt.xlabel('YEARS')
plt.ylabel('ANNUAL RAINFALL (In)')
plt.title('COMPARETIVE RAINFALL IN LAST 18 YEARS')
```

```python
#Sorting out 2017 ANNUAL data
data_rest_pie = data[data.YEAR == 2017]

#CONVERTING NEEDED VALUES TO LISTS
sub = []
annual = []
for item in range(len(data_rest_pie)):
    sub.append(data_rest_pie.SUBDIVISION.iloc[item])
    annual.append(data_rest_pie.ANNUAL.iloc[item])

#CONVERTING INTO DICT FOR SORTING
data_dict = dict(zip( annual , sub ))
data_dict = sorted(data_dict.items() , reverse = True)

#CONVERTING BACK TO LISTS AFTER SOTRING
sub_f = []
annual_f = []
for item in range(len(data_dict)):
    sub_f.append(data_dict[item][1])
    annual_f.append(data_dict[item][0])
```

```python
plt.style.use('seaborn-white')
plt.figure(figsize=(20, 15), dpi= 80)
plt.barh( sub_f , annual_f , color='#2ecc71')

#-----DESIGNING ATTRIBUTES-------------
plt.title('ANNUAL RAINFALL CENSUS (2017)')
plt.grid()
plt.xlabel('ANNUAL RAINFALL (in INCHES)')
plt.tight_layout()
```
