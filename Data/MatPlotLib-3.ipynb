```python
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
slices = [60 , 40 , 50 , 100]
labels = ['Sixty' , 'Fourty' , 'Fifty' , 'Hundred'] 
```

```python
plt.style.use('seaborn-dark')
plt.figure(figsize=(7, 7), dpi= 80)
plt.pie( slices , labels = labels ,
        colors = ['#2ecc71' , '#e74c3c' , '#ffffff' , '#f1c40f' ],
        wedgeprops={'edgecolor':'black' , 'linestyle':'--'})

#----------------------Adding Designes/Attributes-------------------
plt.title('MEDIAN SALARY by AGE')
```

```python
# Language Popularity
popl = [59219, 55466, 47544, 36443, 35917]
lang = ['JavaScript', 'HTML/CSS', 'SQL', 'Python', 'Java']
#EMPHASIZING on One Piece Of Chart BY(EXPLODE)
explode = (0, 0, 0, 0.1 , 0)  # only "explode" the 4th slice (i.e. 'Python')
```

```python
plt.figure(figsize=(7, 7), dpi= 80)
plt.pie( popl , explode = explode ,
        shadow = True , startangle = 90,
        autopct = '%1.1f%%' ,
        labels = lang )
#DESIGN ATTRIBUTES----------------
plt.title('Popularity of Programming Languages')
```
