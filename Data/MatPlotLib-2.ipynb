```python
import pandas as pd
from collections import Counter
data = pd.read_csv(r'DATA/data.csv')

#sorting data and counting the responses
# data.LanguagesWorkedWith[2].split(';')

lang_count = Counter()
for i in range(len(data)):
        lang_count.update(data.LanguagesWorkedWith[i].split(';'))
#lang_count.most_common(15)
```

```python
#sorting the list of tuples to different lists

lang = []
popl = []

for item in lang_count.most_common(15):
    lang.append(item[0])
    popl.append(item[1])
lang.reverse()
popl.reverse()
```

```python
#plotting
plt.style.use('seaborn-white')
plt.figure(figsize=(18, 10), dpi= 80)
plt.barh( lang , popl , color='#2ecc71' )
plt.title('MOST USED PROGRAMMING LANGUAGES')
# plt.ylabel('Programming Languages')
plt.xlabel('Number of People Who Use ')

plt.tight_layout()
```
